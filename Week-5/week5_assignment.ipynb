# Q1. What are the key limitations of traditional MapReduce that make Spark a preferred choice for modern big data processing?

## Answer

### Limitations of MapReduce

- It stores intermediate data on **disk**, which makes processing slow.
- Frequent **disk read and write operations** increase execution time.
- It is **not suitable for iterative tasks** like Machine Learning because it reads data from disk in every iteration.
- Writing MapReduce programs is **more complex** due to separate Map and Reduce functions.
- It mainly supports **batch processing** and has limited built-in features.

### Why Spark is Preferred

- Spark stores data in **memory (RAM)**, making it much faster.
- It uses **lazy evaluation** to optimize execution.
- It supports **SQL, Machine Learning, Streaming, and Graph Processing**.
- Overall, Spark is faster, easier to use, and better suited for modern big data applications.

# Q2. Explain how Spark uses In-Memory Computing to speed up iterative machine learning algorithms compared to disk-based systems.

## Answer

Spark stores intermediate data in **memory (RAM)** instead of writing it to disk after every operation. This reduces disk I/O and makes processing much faster. In machine learning, the same data is used multiple times. Since Spark keeps the data in memory, it avoids reading from disk repeatedly, resulting in faster execution than disk-based systems like MapReduce.

## PySpark Code

```python

# Actions
df.count()
df.show(5)
```

## Explanation

- `cache()` stores the DataFrame in memory.
- The first action loads the data into memory.
- Later operations become faster because Spark uses the cached data instead of reading the file again.

# Q4. Given a DataFrame `df_sales`, write a query to filter for rows where the region is **'West'** and then group by **product_category** to find the average **sale_amount**.

## Answer

This query filters the records where the region is **West** and calculates the average sales for each category.

## Explanation

- `filter()` selects only rows from the West region.
- `groupBy()` groups the data by category.
- `avg()` calculates the average sales for each category.

````python
from pyspark.sql.functions import avg

res = df.filter(df.region == "West") \
        .groupBy("category") \
        .agg(avg("sales").alias("avg_sales"))

res.show()
````

# Q6. Write a query to find the total count of records for each city in a DataFrame, but only for cities where the count is greater than 100.

## Answer

This query counts the number of records in each city and displays only those cities that have more than 100 records.

## Explanation

- `groupBy()` groups the data by city.
- `count()` counts the records in each city.
- `filter()` displays only cities with more than 100 records.

````python
from pyspark.sql.functions import count

res = df.groupBy("region") \
        .agg(count("*").alias("total_records")) \
        .filter("total_records > 100")

res.show()
````

# Q7. How does the immutability of Spark DataFrames affect how you perform data cleaning steps like dropping columns or renaming them?

## Answer

Spark DataFrames are **immutable**, which means they cannot be modified after they are created. Whenever we perform operations like dropping or renaming columns, Spark creates a **new DataFrame** instead of changing the original one.

## Explanation

- The original DataFrame remains unchanged.
- Every transformation returns a new DataFrame.
- This makes data processing safer and more reliable.

````python
# Drop a column
df1 = df.drop("email")

# Rename a column
df2 = df.withColumnRenamed("salary", "monthly_salary")

df1.show(5)
df2.show(5)
````

# Q9. When cleaning a dataset, why is it often better to handle null values before performing mathematical aggregations like `sum()` or `avg()`?

## Answer

Handling null values before performing calculations improves the accuracy of the results. If null values are ignored or left untreated, they may produce incorrect or misleading outputs. Replacing or removing null values ensures reliable aggregations.

## Explanation

- Clean the data before performing calculations.
- Handle null values using `na.fill()` or `na.drop()`.
- This improves the accuracy of `sum()`, `avg()`, and other aggregation functions.

````python
from pyspark.sql.functions import avg

# Fill null salary values with 0
df1 = df.na.fill({"salary": 0})

# Calculate average salary
df1.select(avg("salary").alias("Average Salary")).show()
````

# Q10. Write the code to revise a column named `raw_timestamp` by casting it to a `TimestampType` and renaming it to `event_time`.

## Answer

This query converts a column to the **Timestamp** data type and renames it. In our dataset, we use the `join_date` column instead of `raw_timestamp`.

## Explanation

- `cast()` changes the data type of a column.
- `withColumnRenamed()` renames the column.

````python
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType

df1 = df.withColumn(
    "join_date",
    col("join_date").cast(TimestampType())
)

df1 = df1.withColumnRenamed("join_date", "event_time")

df1.printSchema()
df1.show(5)
````

# Q11. Explain the "Shuffle" process that occurs during a grouping operation. Why is it considered a wide transformation?

## Answer

Shuffle is the process of moving data between different partitions during operations like `groupBy()`, `join()`, and `orderBy()`. It is called a **wide transformation** because data from multiple partitions is exchanged across the cluster before producing the final result.

## Explanation

- Shuffle redistributes data across partitions.
- It occurs during operations like `groupBy()`.
- Since data moves between partitions, it is slower than narrow transformations.

````python
from pyspark.sql.functions import sum

res = df.groupBy("department") \
        .agg(sum("sales").alias("total_sales"))

res.show()
````

# Q12. Write a code snippet that identifies and removes rows where the `email` column contains null values OR the `username` is an empty string.

## Answer

This query removes records where the email is null or the username is empty. Since our dataset does not have a `username` column, we use the `name` column instead.

## Explanation

- `isNull()` checks for null values.
- `filter()` removes unwanted records.
- `~` is used to keep only valid records.

````python
from pyspark.sql.functions import col

df1 = df.filter(
    ~(col("email").isNull() | (col("name") == ""))
)

df1.show(5)
````

# Q13. How do you use the `.agg()` function to calculate multiple statistics at once, such as the `min`, `max`, and `mean` of the `price` column?

## Answer

The `.agg()` function is used to perform multiple aggregate operations in a single query. In our dataset, we use the `salary` column instead of the `price` column.

## Explanation

- `min()` finds the minimum value.
- `max()` finds the maximum value.
- `avg()` calculates the average value.

````python
from pyspark.sql.functions import min, max, avg

df.agg(
    min("salary").alias("Minimum Salary"),
    max("salary").alias("Maximum Salary"),
    avg("salary").alias("Average Salary")
).show()
````

# Q14. In the context of cleaning a dataset, what is the risk of using `inferSchema=true` when your source data contains messy or inconsistent date formats?

## Answer

Using `inferSchema=true` with inconsistent date formats may cause Spark to detect the wrong data type. Some date values may become `null` or be treated as strings, leading to incorrect results during data processing.

## Explanation

- Spark may infer the wrong data type.
- Invalid dates can become `null`.
- It is better to define the schema manually when the data is inconsistent.

````python
from pyspark.sql.types import *

schema = StructType([
    StructField("employee_id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("gender", StringType(), True),
    StructField("department", StringType(), True),
    StructField("region", StringType(), True),
    StructField("category", StringType(), True),
    StructField("salary", IntegerType(), True),
    StructField("sales", IntegerType(), True),
    StructField("experience_years", IntegerType(), True),
    StructField("join_date", DateType(), True),
    StructField("performance_score", DoubleType(), True),
    StructField("email", StringType(), True),
    StructField("status", StringType(), True)
])

df = sp.read.csv(
    "employee.csv",
    header=True,
    schema=schema
)

df.printSchema()
````

# Q15. Write a final processing pipeline that:

- Filters out duplicates.
- Fills null prices with 0.
- Groups by store_id to calculate total revenue.

## Answer

This pipeline cleans the dataset by removing duplicate records, replacing null values with 0, and calculating the total revenue for each department.

## Explanation

- `dropDuplicates()` removes duplicate rows.
- `na.fill()` replaces null values.
- `groupBy()` groups the data by department.
- `sum()` calculates the total revenue.

````python
from pyspark.sql.functions import sum

res = df.dropDuplicates() \
        .na.fill({"sales": 0}) \
        .groupBy("department") \
        .agg(sum("sales").alias("total_revenue"))

res.show()
````
